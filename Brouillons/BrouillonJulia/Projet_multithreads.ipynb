```julia
using Base.Threads
```

```julia
function lire_fichier_csv(thread_id::Int, fichier::String)
    println("Thread $thread_id commence à lire $fichier")
    data = CSV.File(fichier)
    println("Thread $thread_id a fini de lire $fichier")
    return data
end

# Liste des fichiers CSV à lire
fichiers = ["data/expression_data_GSE21653_GSE21653_log_expression_266_samples_21887_genes.csv","data/expression_data_Miller-2005_Miller-2005_log_expression_251_samples_14145_genes.csv",
    "data/expression_data_Naderi-Caldas-2007_Naderi-Caldas-2007_log_expression_242_samples_14366_genes.csv",
    "data/expression_data_probreast_microarrays_E-MTAB-365_log_expression_1190_samples_23035_genes.csv",
    "data/expression_data_probreast_microarrays_GSE25066_log_expression_508_samples_13815_genes.csv",
    "data/expression_data_probreast_microarrays_GSE42568_log_expression_121_samples_23035_genes.csv",
    "data/expression_data_tcga_brca_TCGA-BRCA_log_fpkm_1250_samples_42851_genes.csv",
    "data/expression_data_Yau-2010_Yau-2010_log_expression_683_samples_8791_genes.csv"]

threads = []
for (i, fichier) in enumerate(fichiers)
    thread = Threads.@spawn lire_fichier_csv(i, fichier)
    push!(threads, thread)
end

# Attendez que tous les threads aient terminé
for thread in threads
    Threads.wait(thread)
end
```
